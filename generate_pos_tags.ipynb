```python
import benepar
from IPython.core.debugger import set_trace
import numpy as np
import os

parser = benepar.Parser("benepar_en2_large")

save_dir = "features"
if not os.path.exists(save_dir):
    os.mkdir(save_dir)
```

```python
hp_text = open("chapter9.txt","r").readlines()
all_text = []
para = ""
for i in range(len(hp_text)):
    hp_text[i] = hp_text[i].strip()
    if hp_text[i]!="+":
        para += hp_text[i] + " "
    else:
        all_text.append(para)
        para = ""
```

```python
all_pos_tags = []
for i in range(len(hp_text)):
    all_pos_tags.append([])
    
ind = 0
for para in all_text:
    trees = parser.parse_sents(para)
    for tree in trees:
        treetags = tree.pos()
        rem_len = len(hp_text[ind])
        for tag in treetags:
            all_pos_tags[ind].append(tag[1])
            if tag[0] == "``" or tag[0] == "''" or tag[0] == "-LRB-" or tag[0] == "-RRB-":
                rem_len -= 1
            else:
                rem_len -= len(tag[0])
            if rem_len <= 0:
                ind += 1
                while(ind < len(hp_text) and hp_text[ind] == "+"):
                    ind += 1
                if ind < len(hp_text):
                    rem_len = len(hp_text[ind]) + rem_len

c = [[hp_text[i],all_pos_tags[i]] for i in range(5176)]
for i in c:
    print(i)
```

```python
all_tags = {}
tn = 0
for i in all_pos_tags:
    for j in i:
        if j not in all_tags:
            all_tags[j] = tn
            tn += 1

for i in range(len(all_pos_tags)):
    for j in range(len(all_pos_tags[i])):
        all_pos_tags[i][j] = all_tags[all_pos_tags[i][j]]

print(all_tags)
c = [[hp_text[i],all_pos_tags[i]] for i in range(5176)]
for i in c:
    print(i)
```

```python
import os
one_hot_tags = []
for i in all_pos_tags:
    one_hot_tags.append([0.0]*len(all_tags))
    for j in i:
        one_hot_tags[-1][j] = 1.0

np.save(os.path.join(save_dir,"pos_tags"),one_hot_tags)
np.save(os.path.join(save_dir,"pos_tags_mappings"),all_tags)
```
