```python
import os
import numpy as np
from wordfreq import word_frequency, zipf_frequency, tokenize
from IPython.core.debugger import set_trace

save_dir = "features"
if not os.path.exists(save_dir):
    os.mkdir(save_dir)
```

```python
hp_text = open("chapter9.txt","r").readlines()
all_text = []
para = ""
for i in range(len(hp_text)):
    hp_text[i] = hp_text[i].strip()
```

```python
word_frequencies = []
word_lengths = []

for i in range(len(hp_text)):
    if hp_text[i] == "+":
        word_frequencies.append(0.0)
        word_lengths.append(0)
        continue
    word_frequencies.append(zipf_frequency(hp_text[i], "en", wordlist="large"))
    word_lengths.append(len(hp_text[i]))
```

```python
for i in range(len(hp_text)):
    print("Word: {} Zipf frequency: {} Length: {}".format(hp_text[i], word_frequencies[i], word_lengths[i]))
```

```python
np.save(os.path.join(save_dir,"word_frequency"),word_frequencies)
np.save(os.path.join(save_dir,"word_length"),word_lengths)
```
