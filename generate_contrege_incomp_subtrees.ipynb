```python
import benepar
from IPython.core.debugger import set_trace

parser = benepar.Parser("benepar_en2_large")
```

```python
hp_text = open("chapter9.txt","r").readlines()
all_text = []
para = ""
for i in range(len(hp_text)):
    hp_text[i] = hp_text[i].strip()
    if hp_text[i]!="+":
        para += hp_text[i] + " "
    else:
        all_text.append(para)
        para = ""
```

```python
ex = parser.parse_sents("Harry believed it.")
ex = [i for i in ex]
```

```python
import matplotlib
matplotlib.use('Agg')
print(ex[0].pformat_latex_qtree())
```

```python
all_subtrees = []
for i in range(len(hp_text)):
    all_subtrees.append([])
    
all_lexical_prods = []

ptr = 0
for para in all_text:
    trees = parser.parse_sents(para)
    for tree in trees:
        productions = tree.productions()
        cur_prods = []
        len_remaining = len(hp_text[ptr])
        for prod in productions:
            cur_prods.append(prod)
            if prod.is_lexical():
                if prod.rhs()[0] == "``" or prod.rhs()[0] == "''" or prod.rhs()[0] == "-LRB-" or prod.rhs()[0] == "-RRB-":
                        len_remaining -= 1
                else:
                    len_remaining -= len(prod.rhs()[0])
                all_lexical_prods.append(prod)
            if len_remaining == 0:
                all_subtrees[ptr] = cur_prods.copy()
                ptr += 1
                if hp_text[ptr] == "+":
                    ptr += 1
                if ptr<len(hp_text) and hp_text[ptr] == "+":
                    ptr += 1
                if ptr<len(hp_text):
                    len_remaining = len(hp_text[ptr])
```

```python
for i in range(len(all_subtrees)):
    print(hp_text[i])
    print(all_subtrees[i])
```

```python
sym_to_ind = {}
cnt = 1
all_subgraphs = []

word_to_ind = {}
wordnum = 10000

for subtree in all_subtrees:
    subgraph = []
    for edge in subtree:
        lhs = edge.lhs()
        str_lhs = str(lhs)
        rhs = edge.rhs()
        str_rhs = []
        for r in rhs:
            str_rhs.append(str(r))
        if str_lhs not in sym_to_ind:
            sym_to_ind[str_lhs] = cnt
            cnt += 1
        if edge.is_lexical():
            if id(edge) not in word_to_ind:
                word_to_ind[id(edge)] = wordnum
                wordnum += 1
            subgraph.append("{} {}".format(sym_to_ind[str_lhs],word_to_ind[id(edge)]))
        else:
            for r in str_rhs:
                if r not in sym_to_ind:
                    sym_to_ind[r] = cnt
                    cnt += 1
                subgraph.append("{} {}".format(sym_to_ind[str_lhs],sym_to_ind[r]))
    all_subgraphs.append(subgraph)
```

```python
import os
output_dir = "contrege_incomp_subtrees/"

if not os.path.exists(output_dir):
    os.mkdir(output_dir)

prefix = "subGraph"
for i in range(len(all_subgraphs)):
    if len(all_subgraphs[i]) == 0:
        continue
    with open(output_dir + prefix + str(i),"w+") as f:
        for edge in all_subgraphs[i]:
            f.write(edge + "\n")
```
